```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load EPL data (2023–24 season)
url = "https://www.football-data.co.uk/mmz4281/2324/E0.csv"
df = pd.read_csv(url)

# Rename for easier access
df.rename(columns={
    'FTHG': 'HomeGoals',
    'FTAG': 'AwayGoals',
    'FTR': 'Result'
}, inplace=True)

# Add total goals per match
df['TotalGoals'] = df['HomeGoals'] + df['AwayGoals']

# ----- 1. Top 10 teams by average home goals -----
home_avg = df.groupby('HomeTeam')['HomeGoals'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=home_avg.values, y=home_avg.index, palette='coolwarm')
plt.title("Top 10 Teams by Avg Home Goals (2023–24 EPL)")
plt.xlabel("Average Home Goals")
plt.ylabel("Team")
plt.tight_layout()
plt.show()

# ----- 2. Top 10 teams by average away goals -----
away_avg = df.groupby('AwayTeam')['AwayGoals'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=away_avg.values, y=away_avg.index, palette='crest')
plt.title("Top 10 Teams by Avg Away Goals (2023–24 EPL)")
plt.xlabel("Average Away Goals")
plt.ylabel("Team")
plt.tight_layout()
plt.show()

# ----- 3. Goal Distribution (Histogram) -----
plt.figure(figsize=(8, 4))
sns.histplot(df['TotalGoals'], bins=range(0, 9), kde=False, color='darkgreen')
plt.title("Distribution of Total Goals per Match")
plt.xlabel("Total Goals in Match")
plt.ylabel("Match Count")
plt.xticks(range(0, 9))
plt.tight_layout()
plt.show()

# ----- 4. BTTS (Both Teams To Score) Analysis -----
df['BTTS'] = ((df['HomeGoals'] > 0) & (df['AwayGoals'] > 0))
btts_rate = df['BTTS'].mean() * 100
print(f"BTTS occurred in {btts_rate:.1f}% of matches.")

# ----- 5. Match Result Percentages -----
result_counts = df['Result'].value_counts(normalize=True) * 100

plt.figure(figsize=(6, 6))
plt.pie(result_counts, labels=result_counts.index.map({'H': 'Home Win', 'D': 'Draw', 'A': 'Away Win'}),
        autopct='%1.1f%%', startangle=140, colors=['#2ecc71', '#f1c40f', '#e74c3c'])
plt.title("Match Result Distribution (2023–24 EPL)")
plt.tight_layout()
plt.show()
```
